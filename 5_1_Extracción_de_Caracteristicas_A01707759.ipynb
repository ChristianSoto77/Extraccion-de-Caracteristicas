##**Aplicar métodos para reemplazar valores nulos del dataframe**

```python
MicroRetailer.isnull().sum()
```

```python
#Creamos una copia del dataframe para eliminar los valores no nulos
MicroRetailer1=MicroRetailer.copy()
MicroRetailer1
```

```python
#Procedemos a eliminar los valores nulos con el metodo de promedio y forward fill
Micro_Retailer1=MicroRetailer1.fillna(round(MicroRetailer.mean(),1))
Micro_Retailer1=MicroRetailer1.fillna(method='ffill')
Micro_Retailer1
```

```python
#Corroboramos que ya no existan valores nulos
MicroRetailer1.isnull().sum()
```

##**Aplicar métodos (Desviación Estándar o Rango Intercuartílico) para eliminar Outliers del dataframe**


```python
fig = plt.figure(figsize =(15, 8))
MicroRetailer1.plot(kind='box', vert=False)
plt.title("Distribución de Datos MicroRetailer")
plt.show() 
```

```python
#Método aplicando Cuartiles. Encuentro cuartiles 0.25 y 0.75
y=MicroRetailer1

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido\n", Limite_Superior_iqr)
print()
print("Limite inferior permitido\n", Limite_Inferior_iqr)
```

```python
#Obtenemos datos limpios del Dataframe
outliers_iqr= MicroRetailer1[(y>Limite_Superior_iqr)|(y<Limite_Inferior_iqr)]
outliers_iqr
```

```python
MicroRetailer_clean= MicroRetailer1[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
MicroRetailer_clean
```

Necesitamos buscar un metodo para poder eliminar los valores nulos de las
columnas con información cualitativa y cuantitativa por lo que primero observaremos el tipo de datos de cada variable.

```python
MicroRetailer_str = MicroRetailer_clean.select_dtypes(include=['object'])
MicroRetailer_str.columns
```

```python
MicroRetailer_float = MicroRetailer_clean.select_dtypes(include=['float'])
MicroRetailer_float.columns
```

```python
#Corroboramos valores nulos
MicroRetailer_float.isnull().sum()
```

```python
#Aplicamos el metdo de promedio y ffill a los datos para eliminar valores nulos
cuantitativas=MicroRetailer_float.copy()
cuantitativas=cuantitativas.fillna(round(MicroRetailer_float.mean(),1))
cuantitativas=cuantitativas.fillna(method='ffill')
```

```python
fig = plt.figure(figsize =(15, 8))
cuantitativas.plot(kind='box', vert=False)
plt.title('Nueva distribución de valores de MicroRetailer')
plt.show() #dibujamos el diagrama
```
